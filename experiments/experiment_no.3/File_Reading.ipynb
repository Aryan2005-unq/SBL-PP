To implement a Python program to count the number of lines, words and characters in a file.

```python
if __name__ == '__main__':
  with open('./naruto.txt','r') as file:
    lines = file.readlines()
    words = sum(len(line.split()) for line in lines)
    characters = sum(len(line) for line in lines)
    len_lines = len(lines)


print(f"number of lines are: {len(lines)}")
print(f"number of words are: {words}")
print(f"number of characters are :{characters}")
```
